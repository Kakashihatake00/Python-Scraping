```python
import requests
from bs4 import BeautifulSoup
import csv
proxies = {
    "http" : "http://172.67.188.20:80",
    "http" : "http://172.64.69.38:80",
}
headers = {"User-Agent": "Mozilla/5.0"}
url = "https://search.earth911.com/?what=Electronics&where=1001&list_filter=all&max_distance=100&family_id=&latitude=&longitude=&country=&province=&city=&sponsor="
r = requests.get(url,headers=headers, proxies=proxies)

soup = BeautifulSoup(r.text, "html.parser")
with open("companiesdetail.csv", "w", encoding="utf-8-sig", newline="") as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(["Business_Name","last_update_date", "street_address","materials_accepted"])
    result_container = soup.find("ul",  class_= "result-list")
    print('Scraping started')
    for li in result_container.find_all("li", class_="result-item"):

                link_tag = li.find("a", href=True)
                # print(link_tag)

                detail_url = "https://search.earth911.com" + link_tag['href'] if link_tag else None

                # print(detail_url)

                if detail_url:
                    try:
                        detail = requests.get(detail_url, headers=headers)
                        detail_soup = BeautifulSoup(detail.text, "html.parser")
                        name_tag = detail_soup.find("h1", class_="back-to noprint")
                        name = name_tag.contents[0].strip() if name_tag else "N/A"
                        last_updated_tag = detail_soup.find("span", class_="last-verified")
                        last_updated = last_updated_tag.contents[0].strip() if last_updated_tag else "N/A"
                        location_tag = detail_soup.find("div", class_="contact")
                        # location = location_tag.find("p", class_ = "addr")
                        if location_tag:
                            addresses = location_tag.find_all("p", class_="addr")
                            location = ", ".join([addr.get_text(strip=True) for addr in addresses])
                        else:
                            location = "N/A" 
                        material_tags = detail_soup.find("div", class_="materials-wrapper")
                        if material_tags:
                            material = material_tags.find_all("span", class_="material no-link")
                            count = ", ".join([acc.get_text(strip=True) for acc in material])
                        else:
                            count = "N/A" 
                        # print(name)
                        # print(last_updated)
                        # print(location)
                        # print(count)
                        
                        

                    except Exception as e:
                        print('Error visiting the url')
            

                name = name.replace("ï»¿", "")
                last_updated = last_updated.replace("ï»¿", "")
                location = location.replace("ï»¿", "")
                count = count.replace("ï»¿", "")

                writer.writerow([name, last_updated,location,count])
print("done>>>>")
```
